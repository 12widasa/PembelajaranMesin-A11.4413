```python
import pandas as pd
import re

nama_item = []
harga_item = []

items = soup.find_all("a", "pcv3__info-content css-gwkf0u")
for i in items:
    nama = i.find("div", "prd_link-product-name css-3um8ox").get_text()
    harga = i.find("div","prd_link-product-price css-h66vau").get_text()
    
    # Use regular expression to extract the numeric part of the price
    price_match = re.search(r'(\d+(?:\.\d+)?(?:jt|Jt))', harga)
    if price_match:
        price_str = price_match.group(1).replace('jt', '000').replace('Jt', '000')
        harga_item.append(int(float(price_str)))
        nama_item.append(nama)

dict_items = {"Nama Barang": nama_item, "Harga": harga_item}
df = pd.DataFrame(dict_items, columns = ["Nama Barang", "Harga"])
df
```

```python
import pandas as pd

nama_item = []
harga_item = []

items = soup.find_all("a", "pcv3__info-content css-gwkf0u")

for i in items:
    nama = i.find("div", "prd_link-product-name css-3um8ox").get_text()
    harga = i.find("div","prd_link-product-price css-h66vau").get_text().replace("Rp", "").replace(".", "")
    
    # Split the price range and extract the numeric part
    harga_parts = harga.split(" - ")
    for harga_part in harga_parts:
        harga_part = harga_part.replace("jt", "")
        try:
            harga_num = float(harga_part)
            harga_item.append(int(harga_num * 1))  # Convert to integer
            nama_item.append(nama)
        except ValueError:
            pass

dict_items = {"Nama Barang": nama_item, "Harga": harga_item}
df = pd.DataFrame(dict_items, columns = ["Nama Barang", "Harga"])
df
```
